Getting Tokenize the tweet text

```python
tokenizer = RegexpTokenizer(r'\w+')
DataFrame2["Reviews"] = DataFrame2["Reviews"].apply(tokenizer.tokenize)
```

```python
DataFrame2["Reviews"].head(10)
```

```python
stopwords_list = stopwords.words('arabic')
```

```python
stopwords_list
```

```python
print(len(stopwords_list))
```

```python
listToStr = ' '.join([str(elem) for elem in stopwords_list])
```

```python
listToStr
```

Removing stop words

```python
DataFrame2["Reviews"]=DataFrame2["Reviews"].apply(lambda x: [item for item in x if item not in stopwords_list])
```

Tweet text information

```python
all_words = [word for tokens in DataFrame2["Reviews"] for word in tokens]
sentence_lengths = [len(tokens) for tokens in DataFrame2["Reviews"]]


print("%s words total" % (len(all_words)))
print("Max sentence length is %s" % max(sentence_lengths))
```
