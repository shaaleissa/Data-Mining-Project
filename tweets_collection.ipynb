```python
from ntscraper import Nitter
import pandas as pd
```

```python
scraper = Nitter()
```

```python
tweets= scraper.get_tweets('',to='ZainKSA', language='ar', number=50, near='Dammam')
```

```python
khobar_tweets = scraper.get_tweets('',to='ZainKSA', language='ar', number=100, near='Khobar')
```

```python
hufof_tweets = scraper.get_tweets('',to='ZainKSA', language='ar', number=50, near='ุงููููู')
```

```python
# Initialize an empty list to store the tweet texts
tweet_texts = [tweet['text'] for tweet in tweets['tweets']]

# Convert the list into a DataFrame
dammam_tweets_df = pd.DataFrame(tweet_texts, columns=['Reviews'])

dammam_tweets_df
```

```python
dammam_tweets_df.to_csv('datasets/dammam_tweets.csv', index=False)
```

```python
# Initialize an empty list to store the tweet texts
tweet_texts1 = [khobar_tweet['text'] for khobar_tweet in khobar_tweets['tweets']]

# Convert the list into a DataFrame
khobar_tweets_df = pd.DataFrame(tweet_texts1, columns=['Reviews'])

khobar_tweets_df
```

```python
# Initialize an empty list to store the tweet texts
tweet_texts2 = [hufof_tweet['text'] for hufof_tweet in hufof_tweets['tweets']]

# Convert the list into a DataFrame
hufof_tweets_df = pd.DataFrame(tweet_texts2, columns=['Reviews'])

hufof_tweets_df
```

```python
hufof_tweets_df.to_csv('datasets/hufof_tweets.csv', index=False)
```

```python
khobar_tweets_df.to_csv('datasets/khobar_tweets.csv', index=False)
```
